```python
final_df.columns
```

```python
final_df.groupby(['Aggregate rating','Rating color','Rating text']).size()
```

```python
ratings = final_df.groupby(['Aggregate rating','Rating color','Rating text']).size().reset_index().rename(columns={0:'Rating Count'})
```

```python
ratings
```

### Observation
1. WHEN RATING is [4.5,4.9] -> DARK GREEN -> EXCELLENT
2. WHEN RATING IS [4.0,4.4] -> GREEN      -> VERY GOOD
3. WHEN RATING is [3.5,3.9] -> YELLOW     -> GOOD
4. WHEN RATING is [3.0,3.4] -> ORANGE     -> AVERAGE
5. WHEN RATING is [1.8,2.4] -> RED        -> POOR
6. WHEN RATING is [0.0]     -> WHITE      -> NOT RATED TILL NOW

```python
matplotlib.rcParams['figure.figsize'] = (12,6)
sns.barplot(x="Aggregate rating",y="Rating Count",data=ratings,hue=ratings['Rating color'],palette=['white','red','orange','yellow','green','green'])
```

Observation
1. Not rated count is very high
2. Maximum rating is between 2.5 to 3.4

```python
## Count Plot
sns.countplot(x="Rating color",data=ratings,palette=['blue','red','orange','yellow','green','green'])
```
